```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Storing the data

```python
df=pd.read_csv('Iris.csv')
df.head()
```

# Using The Elbow Method To Find K value

```python
X =df.iloc[:, [0, 1, 2, 3]].values

from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-Cluster Sum of Square') 
```

```python
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

# first two columns

```python

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa',cmap='coolwarm',alpha=0.6)
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour',cmap='coolwarm',alpha=0.6)
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica',cmap='coolwarm',alpha=0.6)


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100,c='black', label = 'Centroids',cmap='rainbow',alpha=0.8)


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
